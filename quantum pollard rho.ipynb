```python
counter = 0
import matplotlib.pyplot as plt
import sys

# Increase the recursion limit to handle complex transpositions, though we will make
# apply_transposition iterative to avoid hitting this limit for its core logic.
sys.setrecursionlimit(5000)

!pip install qiskit
!pip install pylatexenc

from math import ceil, log2

from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import numpy as np
from qiskit.circuit.library import UnitaryGate

# -----------------------------
# Discrete logarithm parameters
# -----------------------------
p = 13
g = 7
h = 10

# Number of qubits required
n = ceil(log2(p))

print("Number of qubits =", n)

# -----------------------------
# Lookup Table
# -----------------------------
lookup_table = {}

for A in range(p - 1):
    for B in range(p - 1):

        h_inverse = pow(h, -B, p)

        value = (pow(g, A, p) * h_inverse) % p

        lookup_table[(A, B)] = value

# print("Lookup Table")
# print("-" * 40)

for (A, B), value in lookup_table.items():
    print(f"A={A}, B={B} -> {value}")

# -----------------------------
# Target State
# -----------------------------
target_value = 1
target_A = None
target_B = None

for (A, B), value in lookup_table.items():

    if value == target_value:
        target_A = A
        target_B = B
        break



# -----------------------------
# Quantum Registers
# -----------------------------
a = QuantumRegister(n, "a")
b = QuantumRegister(n, "b")
y = QuantumRegister(n, "y")      # Function-value register
flag = QuantumRegister(1, "flag")

ca = ClassicalRegister(n, "ca")
cb = ClassicalRegister(n, "cb")
cy = ClassicalRegister(n, "cy")

qc = QuantumCircuit(a, b, y, flag, ca, cb, cy)

# -----------------------------
# Superposition
# -----------------------------
qc.h(a)
qc.h(b)

# Prepare flag in |->
qc.x(flag[0])
qc.h(flag[0])
# ---------- Save Initialization ----------
init_qc = qc.copy()

fig = init_qc.draw(
    output="mpl",
    fold=-1,
    idle_wires=False
)

fig.savefig(
    "Initialization.png",
    dpi=600,
    bbox_inches="tight"
)



# --------------------------------------------------
# Helper Functions (moved to appear before their usage)
# --------------------------------------------------

# Convert an integer into an n-bit binary state
def state_bits(state, n):
    return format(state, f"0{n}b")

# Compare two basis states
def differing_bits(state1, state2):
    bits1 = state1
    bits2 = state2
    positions = []
    for i in range(len(bits1)):
        if bits1[i] != bits2[i]:
            positions.append(i)
    return positions

# Prepare controls for a basis state
def prepare_basis_state(circuit, register, bits):
    n = len(register)
    for i in range(n):
        if bits[n-1-i] == "0":
            circuit.x(register[i])

# Restore basis state
def restore_basis_state(circuit, register, bits):
    n = len(register)
    for i in range(n):
        if bits[n-1-i] == "0":
            circuit.x(register[i])



# Convert one cycle into transpositions
def cycle_to_transpositions(cycle):
    transpositions = []
    if len(cycle) < 2:
        return transpositions
    first = cycle[0]
    # Reverse order
    for i in range(len(cycle) - 1, 0, -1):
        transpositions.append((first, cycle[i]))
    return transpositions

# Multiplication Permutation
def multiplication_permutation(constant, p):
    table = {}
    for x in range(p):
        table[x] = (constant * x) % p
    return table

# Convert permutation into disjoint cycles
def permutation_cycles(table):
    visited = set()
    cycles = []
    for start in table:
        if start in visited:
            continue
        cycle = []
        current = start
        while current not in visited:
            visited.add(current)
            cycle.append(current)
            current = table[current]
        # Ignore fixed points (e.g. 0 -> 0)
        if len(cycle) > 1:
            cycles.append(cycle)
    return cycles

# --------------------------------------------------
# Build Controlled Modular Multiplication Gates
# --------------------------------------------------

gate_cache = {}

def build_modular_gate(constant, p, n):

    size = 2 ** n

    matrix = np.zeros((size, size), dtype=complex)

    for i in range(size):

        if i < p:
            j = (constant * i) % p
        else:
            j = i

        matrix[j][i] = 1

    return UnitaryGate(matrix)


for constant in range(1, p):

    gate_cache[constant] = build_modular_gate(
        constant,
        p,
        n
    )


def controlled_multiply_constant(
        circuit,
        control,
        y,
        constant,
        p
):

    gate = gate_cache[constant]

    controlled_gate = gate.control(1)

    circuit.append(
        controlled_gate,
        [control] + list(y)
    )


# --------------------------------------------------
# Compute Function
# --------------------------------------------------

def compute_function(circuit, a, b, y, p, g, h):

    # --------------------------------------
    # Step 1: Initialize y = 1
    # --------------------------------------
    circuit.x(y[0])

    # --------------------------------------
    # Step 2: Compute g^A
    # --------------------------------------
    for i in range(len(a)):

        constant = pow(g, 2**i, p)

        controlled_multiply_constant(
            circuit,
            a[i],
            y,
            constant,
            p
        )

    # --------------------------------------
    # Step 3: Compute h^(-B)
    # --------------------------------------
    h_inverse = pow(h, -1, p)

    for i in range(len(b)):

        constant = pow(h_inverse, 2**i, p)

        controlled_multiply_constant(
            circuit,
            b[i],
            y,
            constant,
            p
        )
def uncompute_function(circuit, a, b, y, p, g, h):

    # Undo h^{-B}
    h_inverse = pow(h, -1, p)

    for i in reversed(range(len(b))):

        constant = pow(h_inverse, 2**i, p)

        constant_inverse = pow(constant, -1, p)

        controlled_multiply_constant(
            circuit,
            b[i],
            y,
            constant_inverse,
            p
        )

    # Undo g^A
    for i in reversed(range(len(a))):

        constant = pow(g, 2**i, p)

        constant_inverse = pow(constant, -1, p)

        controlled_multiply_constant(
            circuit,
            a[i],
            y,
            constant_inverse,
            p
        )

    # Restore |0001>
    circuit.x(y[0])
# -----------------------------
# Oracle
# -----------------------------
def oracle_y(circuit, y, flag, target_value):

    n = len(y)

    value_bits = format(target_value, f"0{n}b")

    # Convert target value to all-1 controls
    for i in range(n):
        if value_bits[n-1-i] == "0":
            circuit.x(y[i])

    # Mark the state
    circuit.mcx(list(y), flag[0])

    # Restore
    for i in range(n):
        if value_bits[n-1-i] == "0":
            circuit.x(y[i])
# -----------------------------
# Diffusion Operator
# -----------------------------
def diffusion(circuit, qubits):

    circuit.h(qubits)

    circuit.x(qubits)

    circuit.h(qubits[-1])

    circuit.mcx(qubits[:-1], qubits[-1])

    circuit.h(qubits[-1])

    circuit.x(qubits)

    circuit.h(qubits)

# -----------------------------
# Search Register
# -----------------------------
search = list(a) + list(b)

# -----------------------------
# Grover Iterations
# -----------------------------
num_iterations = 1
for _ in range(num_iterations):

    # Compute y
    compute_function(
    qc,
    a,
    b,
    y,
    p,
    g,
    h
)
  # ---------- Save Compute Function ----------
    compute_qc = qc.copy()

    fig = compute_qc.draw(
      output="mpl",
      fold=20,
      idle_wires=False
)

    fig.savefig(
    "Compute_Function.png",
      dpi=600,
      bbox_inches="tight"
)

    # Mark the value
    oracle_y(
        qc,
        y,
        flag,
        target_value
    )
    # ---------- Save Oracle ----------
    oracle_qc = qc.copy()

    fig = oracle_qc.draw(
     output="mpl",
     fold=20,
     idle_wires=False
)

    fig.savefig(
    "Oracle.png",
     dpi=600,
     bbox_inches="tight"
)
    uncompute_function(
    qc,
    a,
    b,
    y,
    p,
    g,
    h
)
    print("\nQuantum function computed.")

    # Uncompute y


    # Diffusion
    diffusion(qc, search)
    # ---------- Save Diffusion ----------
    diffusion_qc = qc.copy()

    fig = diffusion_qc.draw(
     output="mpl",
     fold=20,
     idle_wires=False
)

    fig.savefig(
    "Diffusion.png",
     dpi=600,
     bbox_inches="tight"
)



# -----------------------------
# Measurement
# -----------------------------
qc.measure(a, ca)
qc.measure(b, cb)
qc.measure(y, cy)

# -----------------------------
# Simulation
# -----------------------------
sim = AerSimulator()

compiled = transpile(qc, sim)

result = sim.run(compiled, shots=1024).result()

counts = result.get_counts()

print("\nMeasurement Counts")
print(counts)
print("\nDecoded Results")
print("-" * 40)

import math

valid_results = []

for state, count in counts.items():

    parts = state.split()

    y_bits = parts[0]
    b_bits = parts[1]
    a_bits = parts[2]

    A = int(a_bits, 2)
    B = int(b_bits, 2)

    # Ignore invalid exponents
    if A >= (p - 1):
        continue

    if B >= (p - 1):
        continue

    print(f"A = {A}, B = {B}, Count = {count}")

    valid_results.append((A, B, count))
# -----------------------------
# Histogram
# -----------------------------
plot_histogram(counts)
plt.show()

# -----------------------------
# Circuit
# -----------------------------
# print(qc.draw(output="text"))

# qc.draw("mpl")
# plt.show()
#plt.show()
print(qc.draw(output="text"))

# fig = qc.draw(
#     output="mpl",
#     fold=15,
#     idle_wires=False
# )
try:
    fig = qc.draw(output="mpl")
    fig.savefig("abc.png")
    print("Saved")
except Exception as e:
    print(type(e))
    print(e)
fig.savefig(
    "Complete_Quantum_Pollards_Rho.png",
    dpi=600,
    bbox_inches="tight"
)

print("Circuit saved.")

print("Done")
print("Exists:", os.path.exists("test_circuit.png"))
print(counts)
print("\nVerification")
print("-" * 40)

# Find the most frequent measurement
#best = max(valid_results, key=lambda x: x[2])
invertible_results = []

for A, B, count in valid_results:

    if math.gcd(B, p - 1) == 1:
        invertible_results.append((A, B, count))

if len(invertible_results) == 0:

    print("No invertible B found.")

else:

    best = max(invertible_results, key=lambda x: x[2])

    A = best[0]
    B = best[1]

    print("\nBest Invertible Result")
    print("--------------------")
    print("A =", A)
    print("B =", B)

    B_inverse = pow(B, -1, p - 1)

    x = (A * B_inverse) % (p - 1)

    print("Recovered x =", x)

    print("\nVerification")
    print("--------------------")
    print("g^x mod p =", pow(g, x, p))
    print("Expected h =", h)

    if pow(g, x, p) == h:
        print("\nSUCCESS")
    else:
        print("\nFAILED")

print("\nBest Valid Result")
print("--------------------")
print("A =", A)
print("B =", B)
print("\nRecovering x")
print("--------------------")

modulus = p - 1

if math.gcd(B, modulus) == 1:

    B_inverse = pow(B, -1, modulus)

    x = (A * B_inverse) % modulus

    print("Recovered x =", x)

else:

    print("B is not invertible modulo", modulus)

    x = None

if x is not None:
    value = pow(g, x, p)

    print("\nVerification")
    print("--------------------")
    print("g^x mod p =", value)
    print("Expected h =", h)

    if value == h:
        print("\nSUCCESS")
    else:
        print("\nFAILED")


```
